1. **[1 punto]** Explíquese $L^* = \{\lambda\} \cup L^+$

* El cierre de Kleene de un lenguaje L, que se escribe L*, es el conjunto de todas las cadenas que se pueden formar concatenando elementos de L, y además incluye la cadena vacía. En cambio, el cierre positivo L⁺ es lo mismo pero sin la cadena vacía. Por eso se cumple que L*= {λ} ∪ L⁺
* La equivalencia L*= {λ} ∪ L⁺ se puede explicar a partir de la definición de la clausura positiva. La cerradura positiva de un lenguaje L, es la unión de todas las potencias de L a partir de L^1, es decir, de las concatenaciones de una o más palabras del lenguaje. Por otro lado, la clausura de Kleene incluye también la potencia cero (L^0), que sería exactamente la palabra vacía. Por esta razón, es que L*= {λ} ∪ L⁺

2. **[1 punto]** Explíquese LR $\subset$ LIC

* Los LR son más simples y se pueden describir con AF o ER. Los LIC son más generales y se describen con GIC o AP. Como los LR son un caso particular, siempre están incluidos dentro de los LIC
* Según la jerarquía de Chomsky, podemos afirmar que los LR están contenidos dentro de los LIC ya que clasifica a los LR como un subconjunto de los lenguajes menos restrictivos, como los LIC. Al ser menos restrictivos, las reglas de producción de uno u otro lenguaje son distintas. Por ejemplo, las GR están fuertemente restringidas: en el lado derecho de la producción se limita a un solo terminal, o un no terminal seguido de un terminal (o al revés, dependiendo si es regular por izquierda o por derecha). En cambio, las GIC, dependiendo su formalización, en el lado derecho puede haber uno o más no terminales, intercalados incluso con terminales, siendo así mucho menos restrictivo que las GR.

6. **[1 punto]** Respóndase: Como reconocer un LR infinito si se tiene un número finito de estados en un AF

* Un LR infinito se reconoce porque el AF que lo acepta tiene ciclos en su grafo de estados. Esos ciclos permiten generar cadenas cada vez más largas, ya que se pueden recorrer indefinidamente
* Un LR que es reconocido por un AF y tiene un numero finito de estados, puede ser infinito si el AF tiene algún ciclo desde el cual pueda ser accedido desde el estado inicial, y desde el cual se llegue a un estado final. En particular, si un AF acepta un numero infinito de palabras, entonces debe poder aceptar al menos una palabra que sea lo suficientemente larga como para que se pueda bombear (repetir las veces que se quiera), formando así una cantidad de palabras infinitas y que sigan siendo aceptadas por el AF
