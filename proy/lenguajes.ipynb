## Lenguaje para describir recetas de cocina

* Definir una receta con un nombre
* Agregar ingredientes con cantidades
* Mezclarlos
* Cocinar (hornear) por un tiempo determinado

```
grammar CookLang;

recipe : 'Recipe' STRING '{' step+ '}' ;

step
    : 'Add' quantity ingredient ';'
    | 'Mix' ingredient+ ';'
    | 'Bake' INT 'minutes' ';'
    ;

ingredient : ID ;
quantity   : INT ('g' | 'ml' | 'pcs') ;

ID     : [a-zA-Z]+ ;
INT    : [0-9]+ ;
STRING : '"' .*? '"' ;
WS     : [ \t\r\n]+ -> skip ;
```

```
Recipe "Tarta de manzana" {
    Add 200g harina;
    Add 100g azúcar;
    Add 2pcs manzanas;
    Mix harina azúcar manzanas;
    Bake 45 minutes;
}
```

## Lenguaje para describir flujos de trabajo

* Definir un flujo de trabajo con un nombre
* Especificar tareas y cómo se encadenan
* Incluir decisiones con condiciones que llevan a diferentes caminos

```
grammar FlowLang;

workflow : 'Workflow' STRING '{' task+ '}' ;

task
    : 'Task' STRING '->' STRING ';'
    | 'Decision' STRING '{' option+ '}'
    ;

option
    : 'If' condition 'then' STRING ';'
    ;

condition : ID ;

ID     : [a-zA-Z]+ ;
STRING : '"' .*? '"' ;
WS     : [ \t\r\n]+ -> skip ;
```

```
Workflow "Proceso de contratación" {
    Task "Recibir CV" -> "Revisión inicial";
    Decision "Revisión inicial" {
        If experiencia then "Entrevista técnica";
        If no_experiencia then "Rechazo";
    }
    Task "Entrevista técnica" -> "Oferta";
}
```

## Consulta sobre datos científicos

* Definir un experimento con nombre.
* Medir variables en momentos específicos.
* Registrar valores numéricos.
* Analizar datos con métodos predefinidos (ej. promedio, desviación estándar).

```
grammar LabLang;

experiment : 'Experiment' STRING '{' step+ '}' ;

step
    : 'Measure' STRING 'at' INT 'seconds' ';'
    | 'Record' STRING '=' value ';'
    | 'Analyze' STRING 'with' STRING ';'
    ;

value : INT | FLOAT ;

INT   : [0-9]+ ;
FLOAT : [0-9]+ '.' [0-9]+ ;
STRING: '"' .*? '"' ;
WS    : [ \t\r\n]+ -> skip ;
```

```
Experiment "Reacción química A" {
    Measure "Temperatura" at 10 seconds;
    Record "Temperatura" = 37.5;
    Measure "pH" at 20 seconds;
    Record "pH" = 6.8;
    Analyze "Temperatura" with "Promedio";
}
```

## MiniC

* Definir funciones con parámetros
* Declarar variables con let
* Usar estructuras de control (if, while)
* Operar con expresiones aritméticas

```
grammar MiniC;

program : (function | statement)+ ;

function
    : 'func' ID '(' params? ')' '{' statement+ '}'
    ;

params : ID (',' ID)* ;

statement
    : 'let' ID '=' expr ';'
    | 'if' '(' expr ')' '{' statement+ '}'
    | 'while' '(' expr ')' '{' statement+ '}'
    | 'return' expr ';'
    | expr ';'
    ;

expr
    : INT
    | ID
    | expr '+' expr
    | expr '-' expr
    | expr '*' expr
    | expr '/' expr
    | '(' expr ')'
    ;

ID  : [a-zA-Z]+ ;
INT : [0-9]+ ;
WS  : [ \t\r\n]+ -> skip ;
```

```
func suma(a, b) {
    return a + b;
}

func main() {
    let x = 5;
    let y = 7;
    let z = suma(x, y);
    if (z > 10) {
        return z;
    }
}
```

## Automatización de tareas domésticas

* Definir rutinas con nombre
* Encender y apagar dispositivos
* Configurar valores (ej. temperatura, modo)
* Incluir pausas temporales

```
grammar HomeLang;

routine : 'Routine' STRING '{' action+ '}' ;

action
    : 'TurnOn' device ';'
    | 'TurnOff' device ';'
    | 'Set' device 'to' value ';'
    | 'Wait' INT 'minutes' ';'
    ;

device : ID ;
value  : INT | STRING ;

ID     : [a-zA-Z]+ ;
INT    : [0-9]+ ;
STRING : '"' .*? '"' ;
WS     : [ \t\r\n]+ -> skip ;
```

```
Routine "Mañana en casa" {
    TurnOn luz;
    Set calefaccion to 22;
    Wait 15 minutes;
    TurnOff luz;
}
```
